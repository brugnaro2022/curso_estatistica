Amostragem Aleatória Simples

```python
# Semente, toda vez que roda gera o mesmo resultado
np.random.seed(1)
```

```python
# Pega uma amostragem aleatória, 20%
amos_simples = enem_campinas.sample(frac=0.20)
```

```python
amos_simples
```

```python
# Seleciona aleatóriamente 3000 registros
amos_simples = enem_campinas.sample(3000)
```

```python
amos_simples
```
