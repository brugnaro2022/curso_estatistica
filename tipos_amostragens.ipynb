<h2>Amostragem Estratificada</h2>
<h4>Divisão da população em grupos e seleção aleatória de uma amosra de cada grupo.</h4>

```python
enem_campinas['RACA'].value_counts()
```

<h4>Para quantidade pequena de grupos</h4>

```python
branca = enem_campinas.loc[enem_campinas.RACA == 'branca'].sample(frac=0.15)
branca.shape
```

```python
parda = enem_campinas.loc[enem_campinas.RACA == 'parda'].sample(frac=0.15)
parda.shape
```

```python
preta = enem_campinas.loc[enem_campinas.RACA == 'preta'].sample(frac=0.15)
preta.shape
```

```python
amarela = enem_campinas.loc[enem_campinas.RACA == 'amarela'].sample(frac=0.15)
amarela.shape
```

```python
nd = enem_campinas.loc[enem_campinas.RACA == 'nao_declarado'].sample(frac=0.15)
nd.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, y_treino, amos_estra = train_test_split(
    # extratificando da tabela de dados do enem_campinas um tamanho de teste e 20%, dropando uma parte para ficar com apenas uma das partes
    # axis=1 -> coluna, axis=0 -> linha
    enem_campinas.drop('RACA', axis=1), 
    enem_campinas['RACA'], 
    stratify=enem_campinas['RACA'], 
    test_size=0.20
)
```
